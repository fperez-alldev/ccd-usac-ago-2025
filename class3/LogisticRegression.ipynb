## **Regresi√≥n Log√≠stica**

El algoritmo de **Regresi√≥n Log√≠stica** funciona a partir de la implementaci√≥n de una ecuaci√≥n lineal con variables independientes (tambi√©n llamadas explicativas) para predecir un valor de respuesta.

Por ejemplo, pensemos en la relaci√≥n entre el **n√∫mero de horas estudiadas** y la **probabilidad de aprobar un examen**.

* El n√∫mero de horas estudiadas es la **variable explicativa**, que denotaremos como *x‚ÇÅ*.
* La probabilidad de aprobar el examen es la **variable de respuesta u objetivo**, que denotaremos como *z*.

Si tenemos una sola variable explicativa (*x‚ÇÅ*) y una variable de respuesta (*z*), la ecuaci√≥n lineal se expresa matem√°ticamente como:

```
z = Œ≤‚ÇÄ + Œ≤‚ÇÅx‚ÇÅ
```

Aqu√≠, los coeficientes **Œ≤‚ÇÄ** y **Œ≤‚ÇÅ** son los **par√°metros del modelo**.

---

En caso de tener **m√∫ltiples variables explicativas**, la ecuaci√≥n se generaliza de la siguiente manera:

```
z = Œ≤‚ÇÄ + Œ≤‚ÇÅx‚ÇÅ + Œ≤‚ÇÇx‚ÇÇ + ‚Ä¶ + Œ≤‚Çôx‚Çô
```

En este caso, los coeficientes **Œ≤‚ÇÄ, Œ≤‚ÇÅ, Œ≤‚ÇÇ ‚Ä¶ Œ≤‚Çô** representan los par√°metros del modelo.

---

En conclusi√≥n, el valor de respuesta que se predice (denotado por *z*) se obtiene a trav√©s de estas ecuaciones.


## **Funci√≥n Sigmoide**

El valor de respuesta predicho, representado por *z*, se convierte posteriormente en una **probabilidad** que siempre estar√° entre **0 y 1**.  
Para lograrlo, utilizamos la **funci√≥n sigmoide**, la cual se encarga de transformar cualquier valor real en un valor dentro de ese rango.  

En el campo del *machine learning*, la funci√≥n sigmoide se usa precisamente para **mapear predicciones a probabilidades**.  
Esta funci√≥n tiene la forma de una curva con aspecto de **‚ÄúS‚Äù** suave, por lo que tambi√©n se le conoce como **curva sigmoide**.  

La funci√≥n sigmoide es, en realidad, un caso particular de la **funci√≥n log√≠stica**, y se define matem√°ticamente con la siguiente f√≥rmula:


De manera gr√°fica, la funci√≥n sigmoide puede representarse con una curva en forma de ‚ÄúS‚Äù como se muestra en la siguiente gr√°fica:

### Sigmoid Function

![Sigmoid Function](https://miro.medium.com/max/970/1*Xu7B5y9gp0iL5ooBj7LtWw.png)

## **Frontera de decisi√≥n**

La funci√≥n sigmoide devuelve un valor de **probabilidad** entre 0 y 1.  
Este valor de probabilidad luego se **mapea a una clase discreta**, que puede ser ‚Äú0‚Äù o ‚Äú1‚Äù.  

Para realizar este mapeo (por ejemplo: aprobar/suspender, s√≠/no, verdadero/falso), necesitamos establecer un **umbral**.  
Este umbral se denomina **frontera de decisi√≥n** (*Decision boundary*).  

- Si la probabilidad est√° **por encima del umbral**, clasificamos la observaci√≥n como clase **1**.  
- Si la probabilidad est√° **por debajo del umbral**, la clasificamos como clase **0**.  

---

### Representaci√≥n matem√°tica

$p ‚â• 0.5 => class = 1$

$p < 0.5 => class = 0$

Por lo general, la **frontera de decisi√≥n** se establece en **0.5**.  
* Si el valor de probabilidad es **0.8** (> 0.5), clasificamos la observaci√≥n como **clase 1**.  
* Si el valor de probabilidad es **0.2** (< 0.5), clasificamos la observaci√≥n como **clase 0**.  

Esto puede visualizarse en el siguiente gr√°fico:  


![Decision boundary in sigmoid function](https://ml-cheatsheet.readthedocs.io/en/latest/_images/logistic_regression_sigmoid_w_threshold.png)

## **Realizando predicciones**

Ahora que ya comprendemos la **funci√≥n sigmoide** y la **frontera de decisi√≥n** en la regresi√≥n log√≠stica, podemos usar ese conocimiento para construir una **funci√≥n de predicci√≥n**.  

Una funci√≥n de predicci√≥n en regresi√≥n log√≠stica devuelve la **probabilidad de que una observaci√≥n pertenezca a la clase positiva** (es decir, ‚ÄúS√≠‚Äù o ‚ÄúVerdadero‚Äù).  

A esta clase la llamamos **clase 1**, y se denota como **P(clase = 1)**.  

- Si la probabilidad se acerca m√°s a **1**, tendremos mayor **confianza** en que la observaci√≥n pertenece a la **clase 1**.  
- En caso contrario, la observaci√≥n ser√° clasificada en la **clase 0**.  


# **Supuestos de la Regresi√≥n Log√≠stica**

El modelo de **Regresi√≥n Log√≠stica** requiere cumplir con varios supuestos clave para que sus resultados sean v√°lidos. Estos son los m√°s importantes:

1. La variable dependiente debe ser de tipo **binaria, multinomial u ordinal**.  
   - Ejemplo: aprobar/suspender (binaria), elegir entre varias categor√≠as (multinomial), niveles de satisfacci√≥n (ordinal).

2. Las observaciones deben ser **independientes entre s√≠**.  
   - Es decir, los datos no deben provenir de **mediciones repetidas** sobre el mismo sujeto.

3. Se requiere que exista **poca o ninguna multicolinealidad** entre las variables independientes.  
   - Esto significa que las variables explicativas no deben estar **altamente correlacionadas** entre ellas.

4. El modelo asume **linealidad** entre las variables independientes y los **log-odds** (logaritmo de las probabilidades).

5. El √©xito del modelo depende del **tama√±o de la muestra**.  
   - Por lo general, se necesita un tama√±o de muestra **grande** para lograr una alta precisi√≥n en los resultados.  


## **Tipos de Regresi√≥n Log√≠stica**

El modelo de **Regresi√≥n Log√≠stica** puede clasificarse en tres grupos, seg√∫n las categor√≠as de la variable objetivo. Estos tres grupos se describen a continuaci√≥n:

---

### 1. Regresi√≥n Log√≠stica Binaria

En la **Regresi√≥n Log√≠stica Binaria**, la variable objetivo tiene **dos posibles categor√≠as**.  
Ejemplos comunes de estas categor√≠as son:  
- s√≠ / no  
- bueno / malo  
- verdadero / falso  
- spam / no spam  
- aprobar / suspender  

---

### 2. Regresi√≥n Log√≠stica Multinomial

En la **Regresi√≥n Log√≠stica Multinomial**, la variable objetivo tiene **tres o m√°s categor√≠as**, que no siguen ning√∫n orden espec√≠fico.  
Es decir, existen varias categor√≠as de tipo **nominal**.  

Ejemplo: tipos de frutas ‚Üí manzana, mango, naranja y pl√°tano.  

---

### 3. Regresi√≥n Log√≠stica Ordinal

En la **Regresi√≥n Log√≠stica Ordinal**, la variable objetivo tiene **tres o m√°s categor√≠as con un orden intr√≠nseco**.  
En este caso, las categor√≠as siguen una jerarqu√≠a o secuencia natural.  

Ejemplo: desempe√±o acad√©mico de un estudiante ‚Üí deficiente, regular, bueno, excelente.

### N√∫mero de etiquetas: cardinalidad

El n√∫mero de etiquetas dentro de una variable categ√≥rica se conoce como **cardinalidad**.  
Un alto n√∫mero de etiquetas en una variable se denomina **alta cardinalidad**.  

## ¬øPor qu√© las variables con alta cardinalidad pueden perjudicar un modelo de regresi√≥n log√≠stica?

En **Regresi√≥n Log√≠stica**, las variables categ√≥ricas suelen transformarse en variables num√©ricas mediante t√©cnicas como **One-Hot Encoding**.  
Cuando una variable categ√≥rica tiene **alta cardinalidad** (muchas categor√≠as distintas), esto puede generar varios problemas:

1. **Explosi√≥n en el n√∫mero de variables**  
   - Cada categor√≠a se convierte en una nueva columna.  
   - Con cientos o miles de categor√≠as, el dataset se vuelve muy grande y dif√≠cil de manejar.  

2. **Sobreajuste (*overfitting*)**  
   - El modelo puede ‚Äúmemorizar‚Äù categor√≠as poco frecuentes en lugar de aprender patrones generales.  
   - Esto reduce la capacidad de generalizar en nuevos datos.  

3. **Escasez de datos por categor√≠a**  
   - Algunas categor√≠as aparecen muy pocas veces.  
   - Los coeficientes del modelo resultan poco confiables.  

4. **Mayor complejidad computacional**  
   - El entrenamiento es m√°s lento y costoso.  
   - Aumenta el riesgo de inestabilidad num√©rica.  

---

### Ejemplo
Si una variable `Ciudad` tiene **500 categor√≠as distintas**,  
- con *One-Hot Encoding* se generan **500 columnas adicionales**,  
- y si algunas ciudades tienen solo 2 o 3 registros, los coeficientes no ser√°n representativos.  

---

### Recomendaciones
- **Reducir la cardinalidad** agrupando categor√≠as raras en ‚ÄúOtros‚Äù.  
- Usar **codificaci√≥n alternativa** (*Target Encoding*, *Frequency Encoding*).  
- Evaluar si la variable realmente aporta informaci√≥n √∫til al modelo.

### Supuesto

Asumo que los datos faltantes siguen un patr√≥n **MCAR** (*Missing Completely At Random*), es decir, que se pierden completamente al azar.  

Existen dos m√©todos comunes para imputar valores faltantes:  
1. **Imputaci√≥n con la media o mediana**  
2. **Imputaci√≥n con una muestra aleatoria**  

Cuando el conjunto de datos contiene **valores at√≠picos (outliers)**, es recomendable usar la **mediana**, ya que es m√°s **robusta** frente a valores extremos.  

Por lo tanto, en este caso aplicar√© la **imputaci√≥n con la mediana**.  

---

### Procedimiento
La imputaci√≥n debe hacerse sobre el **conjunto de entrenamiento (train set)** y luego aplicarse al **conjunto de prueba (test set)**.  

Esto significa que las medidas estad√≠sticas utilizadas para rellenar los valores faltantes en ambos conjuntos deben calcularse **√∫nicamente a partir del train set**.

# Imputaci√≥n de datos en variables categor√≠cas

## ¬øPor qu√© usar la moda en la imputaci√≥n de variables categ√≥ricas?

Cuando trabajamos con **variables categ√≥ricas**, los valores faltantes no pueden imputarse con medidas como la media o la mediana, ya que estas solo aplican a datos num√©ricos.  
En su lugar, se utiliza la **moda**, es decir, la categor√≠a que aparece con mayor frecuencia en la variable.

---

### Ventajas de usar la moda
1. **Simplicidad**  
   - Es un m√©todo f√°cil de aplicar y comprender.  

2. **Mantiene categor√≠as v√°lidas**  
   - Los valores faltantes se reemplazan por una categor√≠a existente en los datos, evitando introducir valores artificiales.  

3. **Reduce el sesgo hacia categor√≠as poco frecuentes**  
   - Al usar la categor√≠a m√°s com√∫n, se evita dar demasiado peso a categor√≠as raras que no representan bien el patr√≥n general.  

---

### Limitaciones
- Puede **distorsionar la distribuci√≥n** de la variable si hay muchos valores faltantes.  
- Tiende a **sobrerrepresentar** la categor√≠a m√°s frecuente, lo que puede sesgar el modelo.  

---

### Conclusi√≥n
La **imputaci√≥n por moda** es una t√©cnica apropiada para variables categ√≥ricas, especialmente cuando la cantidad de datos faltantes es peque√±a.  
Si los valores faltantes son numerosos, puede ser mejor explorar m√©todos m√°s avanzados como:  
- **Imputaci√≥n por modelo predictivo**  
- **Imputaci√≥n aleatoria**  
- **Uso de t√©cnicas basadas en √°rboles de decisi√≥n**  


````python
# impute missing categorical variables with most frequent value
for df2 in [X_train, X_test]:
    df2['WindGustDir'].fillna(X_train['WindGustDir'].mode()[0], inplace=True)
    df2['WindDir9am'].fillna(X_train['WindDir9am'].mode()[0], inplace=True)
    df2['WindDir3pm'].fillna(X_train['WindDir3pm'].mode()[0], inplace=True)
    df2['RainToday'].fillna(X_train['RainToday'].mode()[0], inplace=True)
````

### Tratamiento de outliers en variables num√©ricas

Hemos visto que las columnas `Rainfall`, `Evaporation`, `WindSpeed9am` y `WindSpeed3pm` contienen **valores at√≠picos (outliers)**.  
Para tratarlos, utilizar√© el m√©todo de **top-coding**, que consiste en **limitar los valores m√°ximos** con un umbral establecido.  
De esta forma, los valores extremos se reemplazan por el valor m√°ximo permitido, reduciendo el impacto de los outliers en el modelo.  


````python
def max_value(df3, variable, top):
    return np.where(df3[variable]>top, top, df3[variable])

for df3 in [X_train, X_test]:
    df3['Rainfall'] = max_value(df3, 'Rainfall', 3.2)
    df3['Evaporation'] = max_value(df3, 'Evaporation', 21.8)
    df3['WindSpeed9am'] = max_value(df3, 'WindSpeed9am', 55)
    df3['WindSpeed3pm'] = max_value(df3, 'WindSpeed3pm', 57)
````

# **Escalado de caracter√≠sticas (Feature Scaling)**


## ¬øPor qu√© es importante el *Feature Scaling* en Machine Learning?

El **escalado de caracter√≠sticas** (*feature scaling*) es una etapa fundamental en el preprocesamiento de datos para modelos de *machine learning*.  
Consiste en transformar las variables num√©ricas para que se encuentren en **una misma escala** y, as√≠, evitar que unas dominen sobre otras.

---

### üîé Razones principales

1. **Evitar que variables con mayor magnitud dominen el modelo**  
   - Si una variable est√° en miles (ej. ingresos) y otra en unidades (ej. edad), el modelo puede dar m√°s peso a la variable con valores grandes, aunque no sea m√°s relevante.

2. **Mejorar la convergencia de los algoritmos**  
   - Modelos como **Regresi√≥n Log√≠stica, Regresi√≥n Lineal, Redes Neuronales o SVM** usan m√©todos de optimizaci√≥n basados en gradiente.  
   - Si las variables no est√°n escaladas, el entrenamiento puede ser **m√°s lento o inestable**.

3. **Interpretaci√≥n m√°s coherente de los coeficientes**  
   - En modelos lineales, los coeficientes reflejan la relaci√≥n entre cada variable y la predicci√≥n.  
   - Si todas las variables est√°n en la misma escala, la comparaci√≥n de la importancia relativa es m√°s justa.

4. **Mejor desempe√±o en modelos basados en distancia**  
   - Algoritmos como **KNN (K-Nearest Neighbors)** o **K-Means** dependen de c√°lculos de distancia.  
   - Si una variable tiene una escala mucho mayor, dominar√° el c√°lculo de la distancia y sesgar√° los resultados.

---

### ‚öôÔ∏è M√©todos comunes de *Feature Scaling*

1. **Normalizaci√≥n (Min-Max Scaling)**  
   - Escala los valores a un rango fijo, t√≠picamente entre **0 y 1**.  
   - √ötil cuando la distribuci√≥n no es gaussiana.  

   F√≥rmula:  
   $$
   X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}
   $$

2. **Estandarizaci√≥n (Z-Score Scaling)**  
   - Transforma los datos para que tengan **media 0** y **desviaci√≥n est√°ndar 1**.  
   - √ötil cuando los datos tienen una distribuci√≥n aproximadamente normal.  

   F√≥rmula:  
   $$
   X_{std} = \frac{X - \mu}{\sigma}
   $$

---

### ‚úÖ Conclusi√≥n
El **feature scaling** asegura que todas las variables contribuyan de manera equilibrada al modelo,  
mejora la eficiencia de los algoritmos y evita sesgos por diferencias en la magnitud de los datos.

# **Matriz de confusi√≥n**

Una **matriz de confusi√≥n** es una herramienta para **resumir el desempe√±o de un algoritmo de clasificaci√≥n**.  
Nos da una visi√≥n clara del rendimiento del modelo y de los **tipos de errores** que este produce.  

En otras palabras, proporciona un **resumen de las predicciones correctas e incorrectas**, desglosadas por cada categor√≠a, representado en forma tabular.  

---

## Tipos de resultados posibles

Al evaluar un modelo de clasificaci√≥n, existen **cuatro posibles resultados**:

- **Verdaderos Positivos (TP)**  
  Se producen cuando **predecimos** que una observaci√≥n pertenece a cierta clase y, en efecto, la observaci√≥n **pertenece realmente a esa clase**.  

- **Verdaderos Negativos (TN)**  
  Ocurren cuando **predecimos** que una observaci√≥n no pertenece a cierta clase y, efectivamente, la observaci√≥n **no pertenece a esa clase**.  

- **Falsos Positivos (FP)**  
  Se producen cuando **predecimos** que una observaci√≥n pertenece a cierta clase, pero en realidad **no pertenece**.  
  Este tipo de error se denomina **Error Tipo I**.  

- **Falsos Negativos (FN)**  
  Ocurren cuando **predecimos** que una observaci√≥n no pertenece a cierta clase, pero en realidad **s√≠ pertenece**.  
  Este tipo de error es m√°s grave y se denomina **Error Tipo II**.  

---

Estos cuatro resultados se resumen en la siguiente **matriz de confusi√≥n**:  

*(Aqu√≠ ir√≠a la tabla generada por el modelo en el notebook)*  


````python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
````

The confusion matrix shows `20892 + 3285 = 24177 correct predictions` and `3087 + 1175 = 4262 incorrect predictions`.


In this case, we have


- `True Positives` (Actual Positive:1 and Predict Positive:1) - 20892


- `True Negatives` (Actual Negative:0 and Predict Negative:0) - 3285


- `False Positives` (Actual Negative:0 but Predict Positive:1) - 1175 `(Type I error)`


- `False Negatives` (Actual Positive:1 but Predict Negative:0) - 3087 `(Type II error)`

## Precision


**Precision** can be defined as the percentage of correctly predicted positive outcomes out of all the predicted positive outcomes. It can be given as the ratio of true positives (TP) to the sum of true and false positives (TP + FP). 


So, **Precision** identifies the proportion of correctly predicted positive outcome. It is more concerned with the positive class than the negative class.



Mathematically, precision can be defined as the ratio of `TP to (TP + FP).`




````python
# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
````

## Recall


Recall can be defined as the percentage of correctly predicted positive outcomes out of all the actual positive outcomes.
It can be given as the ratio of true positives (TP) to the sum of true positives and false negatives (TP + FN). **Recall** is also called **Sensitivity**.


**Recall** identifies the proportion of correctly predicted actual positives.


Mathematically, recall can be given as the ratio of `TP to (TP + FN).`





````python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
````

## True Positive Rate


**True Positive Rate** is synonymous with **Recall**.


````python
true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
````

## False Positive Rate

````python
false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
````

## Specificity

````python
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
````

## f1-score


**f1-score** is the weighted harmonic mean of precision and recall. The best possible **f1-score** would be 1.0 and the worst 
would be 0.0.  **f1-score** is the harmonic mean of precision and recall. So, **f1-score** is always lower than accuracy measures as they embed precision and recall into their computation. The weighted average of `f1-score` should be used to 
compare classifier models, not global accuracy.
