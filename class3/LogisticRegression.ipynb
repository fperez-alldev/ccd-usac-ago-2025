## **Regresión Logística**

El algoritmo de **Regresión Logística** funciona a partir de la implementación de una ecuación lineal con variables independientes (también llamadas explicativas) para predecir un valor de respuesta.

Por ejemplo, pensemos en la relación entre el **número de horas estudiadas** y la **probabilidad de aprobar un examen**.

* El número de horas estudiadas es la **variable explicativa**, que denotaremos como *x₁*.
* La probabilidad de aprobar el examen es la **variable de respuesta u objetivo**, que denotaremos como *z*.

Si tenemos una sola variable explicativa (*x₁*) y una variable de respuesta (*z*), la ecuación lineal se expresa matemáticamente como:

```
z = β₀ + β₁x₁
```

Aquí, los coeficientes **β₀** y **β₁** son los **parámetros del modelo**.

---

En caso de tener **múltiples variables explicativas**, la ecuación se generaliza de la siguiente manera:

```
z = β₀ + β₁x₁ + β₂x₂ + … + βₙxₙ
```

En este caso, los coeficientes **β₀, β₁, β₂ … βₙ** representan los parámetros del modelo.

---

En conclusión, el valor de respuesta que se predice (denotado por *z*) se obtiene a través de estas ecuaciones.


## **Función Sigmoide**

El valor de respuesta predicho, representado por *z*, se convierte posteriormente en una **probabilidad** que siempre estará entre **0 y 1**.  
Para lograrlo, utilizamos la **función sigmoide**, la cual se encarga de transformar cualquier valor real en un valor dentro de ese rango.  

En el campo del *machine learning*, la función sigmoide se usa precisamente para **mapear predicciones a probabilidades**.  
Esta función tiene la forma de una curva con aspecto de **“S”** suave, por lo que también se le conoce como **curva sigmoide**.  

La función sigmoide es, en realidad, un caso particular de la **función logística**, y se define matemáticamente con la siguiente fórmula:


De manera gráfica, la función sigmoide puede representarse con una curva en forma de “S” como se muestra en la siguiente gráfica:

### Sigmoid Function

## **Frontera de decisión**

La función sigmoide devuelve un valor de **probabilidad** entre 0 y 1.  
Este valor de probabilidad luego se **mapea a una clase discreta**, que puede ser “0” o “1”.  

Para realizar este mapeo (por ejemplo: aprobar/suspender, sí/no, verdadero/falso), necesitamos establecer un **umbral**.  
Este umbral se denomina **frontera de decisión** (*Decision boundary*).  

- Si la probabilidad está **por encima del umbral**, clasificamos la observación como clase **1**.  
- Si la probabilidad está **por debajo del umbral**, la clasificamos como clase **0**.  

---

### Representación matemática

$p ≥ 0.5 => class = 1$

$p < 0.5 => class = 0$

Por lo general, la **frontera de decisión** se establece en **0.5**.  
* Si el valor de probabilidad es **0.8** (> 0.5), clasificamos la observación como **clase 1**.  
* Si el valor de probabilidad es **0.2** (< 0.5), clasificamos la observación como **clase 0**.  

Esto puede visualizarse en el siguiente gráfico:  


![Decision boundary in sigmoid function](https://ml-cheatsheet.readthedocs.io/en/latest/_images/logistic_regression_sigmoid_w_threshold.png)

## **Realizando predicciones**

Ahora que ya comprendemos la **función sigmoide** y la **frontera de decisión** en la regresión logística, podemos usar ese conocimiento para construir una **función de predicción**.  

Una función de predicción en regresión logística devuelve la **probabilidad de que una observación pertenezca a la clase positiva** (es decir, “Sí” o “Verdadero”).  

A esta clase la llamamos **clase 1**, y se denota como **P(clase = 1)**.  

- Si la probabilidad se acerca más a **1**, tendremos mayor **confianza** en que la observación pertenece a la **clase 1**.  
- En caso contrario, la observación será clasificada en la **clase 0**.  


# **Supuestos de la Regresión Logística**

El modelo de **Regresión Logística** requiere cumplir con varios supuestos clave para que sus resultados sean válidos. Estos son los más importantes:

1. La variable dependiente debe ser de tipo **binaria, multinomial u ordinal**.  
   - Ejemplo: aprobar/suspender (binaria), elegir entre varias categorías (multinomial), niveles de satisfacción (ordinal).

2. Las observaciones deben ser **independientes entre sí**.  
   - Es decir, los datos no deben provenir de **mediciones repetidas** sobre el mismo sujeto.

3. Se requiere que exista **poca o ninguna multicolinealidad** entre las variables independientes.  
   - Esto significa que las variables explicativas no deben estar **altamente correlacionadas** entre ellas.

4. El modelo asume **linealidad** entre las variables independientes y los **log-odds** (logaritmo de las probabilidades).

5. El éxito del modelo depende del **tamaño de la muestra**.  
   - Por lo general, se necesita un tamaño de muestra **grande** para lograr una alta precisión en los resultados.  


## **Tipos de Regresión Logística**

El modelo de **Regresión Logística** puede clasificarse en tres grupos, según las categorías de la variable objetivo. Estos tres grupos se describen a continuación:

---

### 1. Regresión Logística Binaria

En la **Regresión Logística Binaria**, la variable objetivo tiene **dos posibles categorías**.  
Ejemplos comunes de estas categorías son:  
- sí / no  
- bueno / malo  
- verdadero / falso  
- spam / no spam  
- aprobar / suspender  

---

### 2. Regresión Logística Multinomial

En la **Regresión Logística Multinomial**, la variable objetivo tiene **tres o más categorías**, que no siguen ningún orden específico.  
Es decir, existen varias categorías de tipo **nominal**.  

Ejemplo: tipos de frutas → manzana, mango, naranja y plátano.  

---

### 3. Regresión Logística Ordinal

En la **Regresión Logística Ordinal**, la variable objetivo tiene **tres o más categorías con un orden intrínseco**.  
En este caso, las categorías siguen una jerarquía o secuencia natural.  

Ejemplo: desempeño académico de un estudiante → deficiente, regular, bueno, excelente.

### Número de etiquetas: cardinalidad

El número de etiquetas dentro de una variable categórica se conoce como **cardinalidad**.  
Un alto número de etiquetas en una variable se denomina **alta cardinalidad**.  

## ¿Por qué las variables con alta cardinalidad pueden perjudicar un modelo de regresión logística?

En **Regresión Logística**, las variables categóricas suelen transformarse en variables numéricas mediante técnicas como **One-Hot Encoding**.  
Cuando una variable categórica tiene **alta cardinalidad** (muchas categorías distintas), esto puede generar varios problemas:

1. **Explosión en el número de variables**  
   - Cada categoría se convierte en una nueva columna.  
   - Con cientos o miles de categorías, el dataset se vuelve muy grande y difícil de manejar.  

2. **Sobreajuste (*overfitting*)**  
   - El modelo puede “memorizar” categorías poco frecuentes en lugar de aprender patrones generales.  
   - Esto reduce la capacidad de generalizar en nuevos datos.  

3. **Escasez de datos por categoría**  
   - Algunas categorías aparecen muy pocas veces.  
   - Los coeficientes del modelo resultan poco confiables.  

4. **Mayor complejidad computacional**  
   - El entrenamiento es más lento y costoso.  
   - Aumenta el riesgo de inestabilidad numérica.  

---

### Ejemplo
Si una variable `Ciudad` tiene **500 categorías distintas**,  
- con *One-Hot Encoding* se generan **500 columnas adicionales**,  
- y si algunas ciudades tienen solo 2 o 3 registros, los coeficientes no serán representativos.  

---

### Recomendaciones
- **Reducir la cardinalidad** agrupando categorías raras en “Otros”.  
- Usar **codificación alternativa** (*Target Encoding*, *Frequency Encoding*).  
- Evaluar si la variable realmente aporta información útil al modelo.

### Supuesto

Asumo que los datos faltantes siguen un patrón **MCAR** (*Missing Completely At Random*), es decir, que se pierden completamente al azar.  

Existen dos métodos comunes para imputar valores faltantes:  
1. **Imputación con la media o mediana**  
2. **Imputación con una muestra aleatoria**  

Cuando el conjunto de datos contiene **valores atípicos (outliers)**, es recomendable usar la **mediana**, ya que es más **robusta** frente a valores extremos.  

Por lo tanto, en este caso aplicaré la **imputación con la mediana**.  

---

### Procedimiento
La imputación debe hacerse sobre el **conjunto de entrenamiento (train set)** y luego aplicarse al **conjunto de prueba (test set)**.  

Esto significa que las medidas estadísticas utilizadas para rellenar los valores faltantes en ambos conjuntos deben calcularse **únicamente a partir del train set**.  

✅ De esta forma se evita el **sobreajuste (overfitting)** y se asegura que el modelo generalice mejor en datos nuevos.  


````python
# impute missing values in X_train and X_test with respective column median in X_train

for df1 in [X_train, X_test]:
    for col in numerical:
        col_median=X_train[col].median()
        df1[col].fillna(col_median, inplace=True)  
````

# Imputación de datos en variables categorícas

## ¿Por qué usar la moda en la imputación de variables categóricas?

Cuando trabajamos con **variables categóricas**, los valores faltantes no pueden imputarse con medidas como la media o la mediana, ya que estas solo aplican a datos numéricos.  
En su lugar, se utiliza la **moda**, es decir, la categoría que aparece con mayor frecuencia en la variable.

---

### Ventajas de usar la moda
1. **Simplicidad**  
   - Es un método fácil de aplicar y comprender.  

2. **Mantiene categorías válidas**  
   - Los valores faltantes se reemplazan por una categoría existente en los datos, evitando introducir valores artificiales.  

3. **Reduce el sesgo hacia categorías poco frecuentes**  
   - Al usar la categoría más común, se evita dar demasiado peso a categorías raras que no representan bien el patrón general.  

---

### Limitaciones
- Puede **distorsionar la distribución** de la variable si hay muchos valores faltantes.  
- Tiende a **sobrerrepresentar** la categoría más frecuente, lo que puede sesgar el modelo.  

---

### Conclusión
La **imputación por moda** es una técnica apropiada para variables categóricas, especialmente cuando la cantidad de datos faltantes es pequeña.  
Si los valores faltantes son numerosos, puede ser mejor explorar métodos más avanzados como:  
- **Imputación por modelo predictivo**  
- **Imputación aleatoria**  
- **Uso de técnicas basadas en árboles de decisión**  


````python
# impute missing categorical variables with most frequent value
for df2 in [X_train, X_test]:
    df2['WindGustDir'].fillna(X_train['WindGustDir'].mode()[0], inplace=True)
    df2['WindDir9am'].fillna(X_train['WindDir9am'].mode()[0], inplace=True)
    df2['WindDir3pm'].fillna(X_train['WindDir3pm'].mode()[0], inplace=True)
    df2['RainToday'].fillna(X_train['RainToday'].mode()[0], inplace=True)
````

### Tratamiento de outliers en variables numéricas

Hemos visto que las columnas `Rainfall`, `Evaporation`, `WindSpeed9am` y `WindSpeed3pm` contienen **valores atípicos (outliers)**.  
Para tratarlos, utilizaré el método de **top-coding**, que consiste en **limitar los valores máximos** con un umbral establecido.  
De esta forma, los valores extremos se reemplazan por el valor máximo permitido, reduciendo el impacto de los outliers en el modelo.  


````python
def max_value(df3, variable, top):
    return np.where(df3[variable]>top, top, df3[variable])

for df3 in [X_train, X_test]:
    df3['Rainfall'] = max_value(df3, 'Rainfall', 3.2)
    df3['Evaporation'] = max_value(df3, 'Evaporation', 21.8)
    df3['WindSpeed9am'] = max_value(df3, 'WindSpeed9am', 55)
    df3['WindSpeed3pm'] = max_value(df3, 'WindSpeed3pm', 57)
````

# **Escalado de características (Feature Scaling)**


## ¿Por qué es importante el *Feature Scaling* en Machine Learning?

El **escalado de características** (*feature scaling*) es una etapa fundamental en el preprocesamiento de datos para modelos de *machine learning*.  
Consiste en transformar las variables numéricas para que se encuentren en **una misma escala** y, así, evitar que unas dominen sobre otras.

---

### 🔎 Razones principales

1. **Evitar que variables con mayor magnitud dominen el modelo**  
   - Si una variable está en miles (ej. ingresos) y otra en unidades (ej. edad), el modelo puede dar más peso a la variable con valores grandes, aunque no sea más relevante.

2. **Mejorar la convergencia de los algoritmos**  
   - Modelos como **Regresión Logística, Regresión Lineal, Redes Neuronales o SVM** usan métodos de optimización basados en gradiente.  
   - Si las variables no están escaladas, el entrenamiento puede ser **más lento o inestable**.

3. **Interpretación más coherente de los coeficientes**  
   - En modelos lineales, los coeficientes reflejan la relación entre cada variable y la predicción.  
   - Si todas las variables están en la misma escala, la comparación de la importancia relativa es más justa.

4. **Mejor desempeño en modelos basados en distancia**  
   - Algoritmos como **KNN (K-Nearest Neighbors)** o **K-Means** dependen de cálculos de distancia.  
   - Si una variable tiene una escala mucho mayor, dominará el cálculo de la distancia y sesgará los resultados.

---

### ⚙️ Métodos comunes de *Feature Scaling*

1. **Normalización (Min-Max Scaling)**  
   - Escala los valores a un rango fijo, típicamente entre **0 y 1**.  
   - Útil cuando la distribución no es gaussiana.  

   Fórmula:  
   $$
   X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}
   $$

2. **Estandarización (Z-Score Scaling)**  
   - Transforma los datos para que tengan **media 0** y **desviación estándar 1**.  
   - Útil cuando los datos tienen una distribución aproximadamente normal.  

   Fórmula:  
   $$
   X_{std} = \frac{X - \mu}{\sigma}
   $$

---

### ✅ Conclusión
El **feature scaling** asegura que todas las variables contribuyan de manera equilibrada al modelo,  
mejora la eficiencia de los algoritmos y evita sesgos por diferencias en la magnitud de los datos.

# **Matriz de confusión**

Una **matriz de confusión** es una herramienta para **resumir el desempeño de un algoritmo de clasificación**.  
Nos da una visión clara del rendimiento del modelo y de los **tipos de errores** que este produce.  

En otras palabras, proporciona un **resumen de las predicciones correctas e incorrectas**, desglosadas por cada categoría, representado en forma tabular.  

---

## Tipos de resultados posibles

Al evaluar un modelo de clasificación, existen **cuatro posibles resultados**:

- **Verdaderos Positivos (TP)**  
  Se producen cuando **predecimos** que una observación pertenece a cierta clase y, en efecto, la observación **pertenece realmente a esa clase**.  

- **Verdaderos Negativos (TN)**  
  Ocurren cuando **predecimos** que una observación no pertenece a cierta clase y, efectivamente, la observación **no pertenece a esa clase**.  

- **Falsos Positivos (FP)**  
  Se producen cuando **predecimos** que una observación pertenece a cierta clase, pero en realidad **no pertenece**.  
  Este tipo de error se denomina **Error Tipo I**.  

- **Falsos Negativos (FN)**  
  Ocurren cuando **predecimos** que una observación no pertenece a cierta clase, pero en realidad **sí pertenece**.  
  Este tipo de error es más grave y se denomina **Error Tipo II**.  

---

Estos cuatro resultados se resumen en la siguiente **matriz de confusión**:  

*(Aquí iría la tabla generada por el modelo en el notebook)*  


````python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
````

The confusion matrix shows `20892 + 3285 = 24177 correct predictions` and `3087 + 1175 = 4262 incorrect predictions`.


In this case, we have


- `True Positives` (Actual Positive:1 and Predict Positive:1) - 20892

- `False Positives` (Actual Negative:0 but Predict Positive:1) - 1175 `(Type I error)`


- `False Negatives` (Actual Positive:1 but Predict Negative:0) - 3087 `(Type II error)`

## Precision


**Precision** can be defined as the percentage of correctly predicted positive outcomes out of all the predicted positive outcomes. It can be given as the ratio of true positives (TP) to the sum of true and false positives (TP + FP). 


So, **Precision** identifies the proportion of correctly predicted positive outcome. It is more concerned with the positive class than the negative class.



Mathematically, precision can be defined as the ratio of `TP to (TP + FP).`




````python
# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
````

## Recall


Recall can be defined as the percentage of correctly predicted positive outcomes out of all the actual positive outcomes.
It can be given as the ratio of true positives (TP) to the sum of true positives and false negatives (TP + FN). **Recall** is also called **Sensitivity**.


**Recall** identifies the proportion of correctly predicted actual positives.


Mathematically, recall can be given as the ratio of `TP to (TP + FN).`





````python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
````

## True Positive Rate


**True Positive Rate** is synonymous with **Recall**.


````python
true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
````

## False Positive Rate

````python
false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
````

## Specificity

````python
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
````

## f1-score


**f1-score** is the weighted harmonic mean of precision and recall. The best possible **f1-score** would be 1.0 and the worst 
would be 0.0.  **f1-score** is the harmonic mean of precision and recall. So, **f1-score** is always lower than accuracy measures as they embed precision and recall into their computation. The weighted average of `f1-score` should be used to 
compare classifier models, not global accuracy.
